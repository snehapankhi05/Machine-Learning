```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
titantic=pd.read_csv('train.csv')
titantic
```

```python
flights=sns.load_dataset('flights')
flights
```

```python
iris=sns.load_dataset('iris')
iris
```

# 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,hue=tips['sex'],style=tips['smoker'],size=tips['size'])
plt.title('Total Bill vs Tip')
plt.show()
```

# 2 Barplot (Numerical-Categorical)

```python
titantic.head()
```

```python
sns.barplot(x='Pclass', y='Fare', data=titantic, hue=titantic['Sex'])
plt.title('Average Age by Passenger Class')
plt.show()
```

# 3 Box Plot (Numerical - Categorical)

```python
sns.boxplot(x='Sex', y='Age', data=titantic, hue=titantic['Survived'])
```

# 4 Distplot (Numerical- Categorical)

```python
sns.histplot(titantic[titantic['Survived']==0]['Age'],kde=True)
sns.histplot(titantic[titantic['Survived']==1]['Age'],kde=True)
```

# 5 Heatmap (Categorical-Categorical)

```python
sns.heatmap(pd.crosstab(titantic['Pclass'],titantic['Survived']))
```

```python
survival_rate_by_sex = titantic.groupby('Sex')['Survived'].mean() * 100
survival_rate_by_sex
```

# 6 ClusterMap(Categorical-Categorical)

```python
sns.clustermap(pd.crosstab(titantic['SibSp'],titantic['Survived']))
```

# 7 Pairplot

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

# 8 Lineplot (Numerical-Numerical)

```python
flights.head()
```

```python
flights_yearly = flights.groupby('year')['passengers'].sum().reset_index()
flights_yearly
```

```python
sns.lineplot(x=flights_yearly['year'], y=flights_yearly['passengers'])
plt.title("Total Passengers per Year")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.show()
```
