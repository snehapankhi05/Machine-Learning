```python
import pandas as pd
import seaborn as sns
df=pd.read_csv('train.csv')
df.head()
```

# 1 Categorical Data

## a. Countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# for Survived column
sns.countplot(x='Survived', data=df,)
plt.show()

```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
# for Pclass
sns.countplot(
    x='Pclass',
    hue='Pclass',
    data=df,
    palette=['#ff4d6d', '#0077b6','orange'],
    legend=False
)
plt.show()
```

```python
df['Pclass'].value_counts()
```

```python
# sex column
sns.countplot(x='Sex', data=df)
plt.show()
```

```python
df['Sex'].value_counts()
```

## b.Piechart

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2 Numerical Data

## a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])
```

## b. Distplot

```python
sns.histplot(df['Age'], kde=True, stat='probability', color='orchid', bins=25, edgecolor='white', alpha=0.8)
plt.title('Age Distribution (Probability)', fontsize=13)
plt.ylabel('Probability')
plt.xlabel('Age')
plt.show()

#kde--> adds smooth line
```

```python
# skkewness
df['Age'].skew()
```

## c. Boxplot

```python
# Boxplot of fare

sns.boxplot(x=df['Fare'], color='#ff4d6d', width=0.4)
plt.title('Boxplot of Fare', fontsize=13)
plt.show()
```

```python
# boxplot of Age
sns.boxplot(x=df['Age'], color='red', width=0.4)
plt.title('Boxplot of Age', fontsize=13)
plt.show()
```
